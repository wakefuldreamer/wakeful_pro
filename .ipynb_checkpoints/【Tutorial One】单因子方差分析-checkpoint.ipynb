åœ¨ä¸Šå›¾ä¸­ï¼Œå¯ä»¥ä»ç›´è§‚ä¸Šæ¯”è¾ƒä¸¤ç§é¥®é£Ÿæ–¹å¼å¯¹ä½“é‡å˜åŒ–é‡çš„å½±å“ã€‚å¯¹äºé‡‡ç”¨é¥®é£Ÿæ–¹å¼Aï¼Œå¿—æ„¿è€…çš„ä½“é‡å˜åŒ–é‡é›†ä¸­åœ¨$3$é™„è¿‘ï¼Œä½†æœ‰ä¸¤ä¸ªè¾ƒå¤§çš„æ•°å€¼ï¼›å¯¹äºé‡‡ç”¨é¥®é£Ÿæ–¹å¼Bï¼Œå¿—æ„¿è€…çš„ä½“é‡å˜åŒ–é‡é›†ä¸­åœ¨$4$é™„è¿‘ï¼Œä½†ä¹Ÿæœ‰ä¸‰ä¸ªè´Ÿå€¼ã€‚æ„Ÿè§‰ä¸Šï¼Œä¸¤ç§é¥®é£Ÿæ–¹å¼æ˜¯æœ‰åŒºåˆ«çš„ã€‚ä½†æ˜¯ï¼Œè¿™ç§â€œåŒºåˆ«â€ä¼šå¦æ˜¯ç”±äºå¿—æ„¿è€…ä¸ªä½“å·®å¼‚æ‰€é€ æˆçš„ï¼Ÿè¿˜æ˜¯è¿™ç§â€œåŒºåˆ«â€æ˜¯æ˜æ˜¾åœ°æœ‰åˆ«äºä¸ªä½“å·®å¼‚çš„ï¼Ÿæˆ‘ä»¬ä»éœ€è¦ç”¨é‡åŒ–çš„æ‰‹æ®µåŠ ä»¥è¡¡é‡ã€‚ä¸ºäº†æ¯”è¾ƒä¸¤ç»„æ•°æ®æ˜¯å¦å­˜åœ¨å·®å¼‚ï¼Œä¸€ç§å¸¸ç”¨çš„æ–¹æ³•æ˜¯äºŒæ ·æœ¬ç‹¬ç«‹$t$æ£€éªŒã€‚

æ¥ä¸‹æ¥ï¼Œæˆ‘ä»¬å›é¡¾ä¸€ä¸‹äºŒæ ·æœ¬ç‹¬ç«‹$t$æ£€éªŒã€‚

å‡å®šç¬¬ä¸€ç»„æ•°æ®$x_1,x_2,\cdots,x_{m_1}$æ¥è‡ªä¸€ä¸ªæ­£æ€åˆ†å¸ƒ$N(\mu_1,\sigma^2)$ï¼Œç¬¬äºŒç»„æ•°æ®$y_1,y_2,\cdots,y_{m_2}$æ¥è‡ªå¦ä¸€ä¸ªæ­£æ€åˆ†å¸ƒ$N(\mu_2,\sigma^2)$ï¼Œå¹¶ä¸”è¿™ä¸¤ä¸ªæ­£æ€åˆ†å¸ƒæ˜¯ç›¸äº’ç‹¬ç«‹çš„ã€‚äºæ˜¯ï¼Œæ¯”è¾ƒä¸¤ç»„æ•°æ®å·®å¼‚çš„é—®é¢˜ï¼Œå°±è½¬åŒ–ä¸ºæ¯”è¾ƒä¸¤ä¸ªæ€»ä½“å‡å€¼å·®å¼‚çš„é—®é¢˜ã€‚é€šå¸¸ï¼Œæˆ‘ä»¬ä¼šè€ƒè™‘è¿™æ ·çš„ä¸€å¯¹å‡è®¾ï¼Œå³åŸå‡è®¾å’Œå¤‡æ‹©å‡è®¾åˆ†åˆ«ä¸º
$$
H_0: \mu_1 = \mu_2, \quad \text{vs} \quad H_1: \mu_1\neq \mu_2.
$$
ä¸ºè§£å†³è¿™ä¸ªé—®é¢˜ï¼Œæˆ‘ä»¬æ„å»ºäº†ä¸€ä¸ªæ£€éªŒç»Ÿè®¡é‡ä¸º
$$
t_0 = \frac{ (\bar{x} - \bar{y})}{\sqrt{s_w^2 ({m_1}^{-1} + {m_2}^{-1})}}.
$$
å…¶ä¸­ï¼Œ
- $\bar{x} = {m_1}^{-1} \sum_{i=1}^{m_1} x_i$è¡¨ç¤ºç¬¬ä¸€ç»„çš„æ ·æœ¬å‡å€¼ï¼›
- $\bar{y} = {m_2}^{-1} \sum_{i=1}^{m_2} y_i$è¡¨ç¤ºç¬¬äºŒç»„çš„æ ·æœ¬å‡å€¼ï¼›
- $s_w^2 = \frac{(m_1-1)\sum{i=1}^{m_1}(x_i-\bar{x})^2 + (m_2-1)\sum_{j=1}^{m_2} (y_j-\bar{y})^2}{m_1+m_2-2}$è¡¨ç¤ºä¸¤ç»„æ ·æœ¬çš„åˆæ–¹å·®ã€‚

å¯ä»¥è¯æ˜ï¼Œåœ¨åŸå‡è®¾æˆç«‹çš„æƒ…å†µä¸‹ï¼Œ$t \sim t(m_1+m_2-2)$ã€‚äºæ˜¯ï¼Œæˆ‘ä»¬å¯ä»¥æ„é€ æ‹’ç»åŸŸä¸º$\{|t_0| > t_{1-\alpha/2}(m_1+m_2-2)\}$ã€‚å½“æ ·æœ¬è½å…¥æ‹’ç»åŸŸæ—¶ï¼Œåˆ™æ‹’ç»åŸå‡è®¾ï¼›åä¹‹ï¼Œåˆ™æ— æ³•æ‹’ç»åŸå‡è®¾ã€‚é™¤äº†æ„é€ æ‹’ç»åŸŸè¿™ä¸€æ–¹æ³•ä¹‹å¤–ï¼Œæˆ‘ä»¬è¿˜å¯ä»¥è®¡ç®—å…¶$p$å€¼ï¼Œå³
$$
p = 2\times P(t_{m_1+m_2-2} > |t_0|) .
$$
å¦‚æœ$p$å€¼å°äºæ˜¾è‘—æ€§æ°´å¹³$\alpha$ï¼Œåˆ™æ‹’ç»åŸå‡è®¾ï¼›åä¹‹ï¼Œåˆ™æ— æ³•æ‹’ç»åŸå‡è®¾ã€‚

æ³¨ï¼šåœ¨è¿™ä¸ªé—®é¢˜ä¸­ï¼Œ$m_1=m_2=24$ã€‚

````python
# Reconstruct a dataset
Data2 = Data1.values
list_type = ["A","B"]
Group2_data = [Data2[Data2[:,0] == x,1] for x in list_type]
# Two-sample t test
t0,pVal0 = stats.ttest_ind(Group2_data[0],Group2_data[1])
````

é¦–å…ˆï¼Œæˆ‘ä»¬å¯ä»¥è®¡ç®— $t$ æ£€éªŒç»Ÿè®¡é‡ä¸º

````python
print("t statistic is:" ,round(t0,4))
````

ä¸å…¶ç›¸åº”çš„ä¸´ç•Œå€¼è¿›è¡Œæ¯”è¾ƒï¼Œå³$t_{1-\alpha/2}(m+n-2)$ä¸º

````python
print("critical value isï¼š", round(t.ppf(1-alpha/2,24*2-2),4))
````

æˆ‘ä»¬å‘ç°ï¼Œ$t$ æ£€éªŒç»Ÿè®¡é‡çš„ç»å¯¹å€¼æ¯”ä¸´ç•Œå€¼å°å¾—å¤šï¼Œæ‰€ä»¥ï¼Œæ— æ³•æ‹’ç»åŸå‡è®¾ï¼Œä¹Ÿå°±æ˜¯è¯´ï¼Œè¿™ä¸¤ç§é¥®é£Ÿæ–¹å¼å¯¹ä½“é‡çš„å˜åŒ–çš„å½±å“æ— æ˜¾è‘—å·®å¼‚ã€‚

é™¤äº†ä¸Šè¿°æ–¹æ³•ä¹‹å¤–ï¼Œæˆ‘ä»¬ä»å¯ä»¥é‡‡ç”¨ $p$ å€¼æ¥å¾—åˆ°ç»“è®ºã€‚å…·ä½“æ¥è¯´ï¼Œæˆ‘ä»¬å¯ä»¥è®¡ç®— $p$ å€¼ä¸º

````python
print("p-value is:", round(pVal0,4))
````

### Task2: åˆ©ç”¨å•å› ç´ æ–¹å·®åˆ†ææ¨¡å‹ï¼Œæ¯”è¾ƒä¸‰ç§é¥®é£Ÿæ–¹å¼å¯¹å‡é‡çš„å½±å“æ˜¯å¦å­˜åœ¨å·®å¼‚ï¼Ÿ

ä¸Q1ç±»ä¼¼ï¼ŒQ2æœ¬è´¨ä¸Šå¯è½¬åŒ–ä¸ºï¼Œå¯¹äºé‡‡ç”¨ä¸åŒé¥®é£Ÿæ–¹å¼çš„å¿—æ„¿è€…ï¼Œä»–ä»¬çš„ä½“é‡å˜åŒ–é‡åˆ†å¸ƒæ˜¯å¦å­˜åœ¨å·®å¼‚ï¼Ÿå¯¹äº $a=2$ ç»„æ ·æœ¬çš„å‡å€¼æ¯”è¾ƒï¼Œæœ€ä¸ºå¸¸ç”¨çš„æ–¹æ³•æ˜¯ $t$ æ£€éªŒã€‚ç„¶è€Œï¼Œå¯¹äº$a\geq 2$æ—¶ï¼Œæ›´ä¸ºä¸€èˆ¬çš„æ–¹æ³•æ˜¯**å•å› å­æ–¹å·®åˆ†ææ¨¡å‹**ã€‚

è¿™é‡Œæˆ‘ä»¬ç®€å•ä»‹ç»ä¸€ä¸‹å•å› ç´ æ–¹å·®åˆ†ææ¨¡å‹ï¼ˆOne-way ANOVA æ¨¡å‹ï¼‰ã€‚

å‡è®¾æœ‰ $a$ ç»„æ•°æ®ï¼Œæ¯ç»„æ•°æ®æœ‰ $m$ ä¸ªï¼Œäºæ˜¯ï¼Œå…±æœ‰ $n = am$ ä¸ªæ•°æ®ã€‚å¯¹äºç¬¬$i$ç»„çš„æ•°æ®è€Œè¨€ï¼Œæˆ‘ä»¬å°†æ•°æ®è®°ä¸º$y_{i1},y_{i2},\cdots,y_{ij},\cdots,y_{im}$ã€‚
åœ¨æ–¹å·®åˆ†ææ¨¡å‹ä¸­ï¼Œæˆ‘ä»¬å‡å®š:

ï¼ˆ1ï¼‰æ¯ä¸€ä¸ªæ•°æ® $y_{ij} = \mu + \alpha_i + \epsilon_{ij}$ï¼Œå…¶ä¸­ï¼Œ$\mu$æ˜¯æœªçŸ¥å‚æ•°ï¼Œè¡¨ç¤ºæ€»ä½“å‡å€¼ï¼Œä¸ç»„åˆ«å’Œæ•°æ®å·®å¼‚æ— å…³ï¼›$\alpha_i$æ˜¯æœªçŸ¥å‚æ•°ï¼Œè¡¨ç¤ºç»„å†…å‡å€¼ï¼Œä¸æ•°æ®å·®å¼‚æ— å…³ï¼Œ$\sum_{i=1}^a\alpha_i = 0$ï¼›$\epsilon_{ij}$æ˜¯éšæœºè¯¯å·®ï¼Œè¡¨ç¤ºæ•°æ®ä¹‹é—´çš„å·®å¼‚ï¼›

ï¼ˆ2ï¼‰$\epsilon_{ij}\overset{i.i.d}{\sim}N(0,\sigma^2)$ï¼Œ$\sigma^2$è¡¨ç¤ºæ•°æ®çš„æ³¢åŠ¨ã€‚

æ ¹æ®ä¸Šè¿°å‡å®šï¼Œæ˜“çŸ¥
$$
y_{ij} \sim N(\mu+\alpha_i , \sigma^2) \quad \text{$x_{ij}$ ä¹‹é—´ç›¸äº’ç‹¬ç«‹ã€‚}
$$
åŸºäºä¸Šè¿°å‡å®šï¼Œâ€œæ•°æ®åˆ†å¸ƒä¸€è‡´çš„é—®é¢˜â€å¯ä»¥è½¬åŒ–ä¸ºâ€œå‡å€¼æ˜¯å¦ç›¸ç­‰çš„é—®é¢˜â€ã€‚æˆ‘ä»¬æ„å»ºäº†ä¸€å¯¹å‡è®¾ï¼Œåˆ†åˆ«ä¸º
$$
H_0: \alpha_1 = \alpha_2 = \alpha_3 = 0 \quad \text{vs} \quad H_1: \text{$\alpha_i$ ä¸å…¨ç›¸ç­‰ã€‚}
$$

åœ¨æ–¹å·®åˆ†ææ¨¡å‹ä¸­ï¼Œæˆ‘ä»¬éœ€è¦å»ºç«‹æ–¹å·®åˆ†æè¡¨ï¼Œå³

| æ¥æº  | è‡ªç”±åº¦ | å¹³æ–¹å’Œ  | å‡æ–¹ | ç»Ÿè®¡é‡ | 
| ----  | ----- |  ----- | ----- | ----- |
| å› å­  | a-1 | SSA | MSA | F |
| è¯¯å·® | n-a | SSE | MSE |  |
| æ€»å’Œ | n-1 | SST |    |  |

å…¶ä¸­ï¼Œ
- $SS_A = m\sum_{i=1}^{a} (\bar{y}_{i\cdot}-\bar{y}_{\cdot\cdot})^2$;
- $MS_A = (a-1)^{-1}SS_A$;
- $SS_E = \sum_{i=1}^{a} \sum_{j=1}^m (y_{ij} - \bar{y}_{i\cdot})^2$;
- $MS_E = (n-a)^{-1}SS_E $;
- $F = MS_A/MS_E$ã€‚

æˆ‘ä»¬å›åˆ°è¿™ä¸ªæ•°æ®é›†ã€‚æ¯”è¾ƒåœ¨ä¸‰ç§é¥®é£Ÿæ–¹å¼ä¸‹ï¼Œå¿—æ„¿è€…ä½“é‡å˜åŒ–é‡çš„å¹³å‡æ°´å¹³æ˜¯å¦ä¸€è‡´ï¼Ÿæ„å»ºANOVAè¡¨å¦‚ä¸‹ï¼š

````python
model = ols('Weight_loss~Diet_type', Data).fit() # Add C() to numeric group indices
anovaResults = round(anova_lm(model),4)
print("\nThe ANOVA table: \n", anovaResults)
````

### Task 3: ç»™å‡ºå•å› ç´ æ–¹å·®åˆ†ææ¨¡å‹ä¸­å‚æ•°çš„ä¼°è®¡å€¼ï¼Œå¹¶é˜è¿°å…¶å«ä¹‰ã€‚

- $\mu$: æ€»ä½“å¹³å‡æ°´å¹³ï¼›
- $\alpha_i$: ç¬¬$i$ä¸ªæ°´å¹³çš„æ•ˆåº”å€¼ï¼›
- $\sigma^2$: è¯¯å·®æ–¹å·®ã€‚

$\mu$çš„æœ€å¤§ä¼¼ç„¶ä¼°è®¡æ˜¯
$$
\hat{\mu} = \bar{y}_{\cdot\cdot} = \frac{1}{am}\sum_{i=1}^a\sum_{j=1}^m y_{ij};
$$

$\alpha_i$çš„æœ€å¤§ä¼¼ç„¶ä¼°è®¡æ˜¯
$$
\hat{\alpha}_i = \bar{y}_{i\cdot} - \bar{y}_{\cdot\cdot} = \frac{1}{m}\sum_{j=1}^m y_{ij} - \frac{1}{am}\sum_{i=1}^a\sum_{j=1}^m y_{ij};
$$

$\sigma^2$çš„å¸¸ç”¨ä¼°è®¡æ˜¯
$$
\hat{\sigma}^2 = \frac{SS_E}{a(m-1)} = \frac{\sum_{i=1}^a\sum_{j=1}^m (y_{ij} - \bar{y}_{i\cdot})^2}{n-a}= MS_E;
$$

äºæ˜¯ï¼Œæˆ‘ä»¬å¯ä»¥è®¡ç®—

$$
e_{i} = \rho e_{i-1} + v_{i}, \quad v_i \overset{i.i.d}{\sim} N(0,\sigma_v^2)ï¼Œ
$$
å…¶ä¸­ï¼Œ$|\rho| < 1$ã€‚

åœ¨Durbin-Watsonæ£€éªŒä¸­ï¼Œæˆ‘ä»¬éœ€è¦æ£€éªŒè¿™æ ·ä¸€å¯¹å‡è®¾
$$

å¯ä»¥æ³¨æ„åˆ°ï¼Œ$\rho = 0$è¡¨ç¤ºæ®‹å·®åºåˆ—æ— è‡ªç›¸å…³æ€§ã€‚ä¹Ÿå°±æ˜¯è¯´ï¼Œå¦‚æœæˆ‘ä»¬æ¥å—åŸå‡è®¾ï¼Œé‚£ä¹ˆæˆ‘ä»¬å°±è®¤ä¸ºåºåˆ—æ•°æ®æ˜¯æ»¡è¶³ç‹¬ç«‹æ€§çš„ã€‚

DWæ£€éªŒç»Ÿè®¡é‡ä¸º
$$
DW = \frac{\sum_{i=2}^n (e_i - e_{i-1})^2}{\sum_{i=1}^n e_{i}^2}
$$
å¯ä»¥æ³¨æ„åˆ°ï¼Œ$DW \approx 2(1-r_1)$ï¼Œå…¶ä¸­$r_1$è¡¨ç¤ºä¸€é˜¶æ»åæ ·æœ¬è‡ªç›¸å…³ç³»æ•°ï¼Œå³
$$

$$
å¯¹äºæ— ç›¸å…³æ€§çš„æ®‹å·®åºåˆ—ï¼Œ$r_1$æ¥è¿‘äºé›¶ï¼Œæ‰€ä»¥$DW$ç»Ÿè®¡é‡æ¥è¿‘äº$2$ã€‚

- å¦‚æœ $d_L \leq DW \leq d_U$, é‚£ä¹ˆæ— æ³•åˆ¤æ–­ã€‚

$d_L$å’Œ$d_U$ä¸æ ·æœ¬é‡ã€çº¿æ€§æ¨¡å‹ä¸­ç‰¹å¾ä¸ªæ•°ä»¥åŠæ˜¾è‘—æ€§æ°´å¹³æœ‰å…³ã€‚å…·ä½“å¯ä»¥æŸ¥è¡¨ï¼Œå¦‚ä¸‹ï¼š![DW_table.jpg](Figures/DW_table.jpg)

##### Bartlettæ£€éªŒ
åœ¨Bartlettæ£€éªŒä¸­ï¼Œæ‰€ä½¿ç”¨çš„æ£€éªŒç»Ÿè®¡é‡ä¸º
$$
\chi_0^2 = 2.3026 \frac{q}{c}
$$
å…¶ä¸­ï¼Œ
- $ q = (n - a) \log_{10}s_p^2 - \sum_{i=1}^a (m_i-1)  \log_{10}s_i^2 $ï¼›
- $c = 1 + \frac{1}{3(a-1)}\left( \sum_{i=1}^a (m_i-1)^{-1} - (n-a)^{-1} \right)$ï¼›
- $s_p^2 = \frac{\sum_{i}  (m_i-1)s^2_i}{n-a}$ï¼›
- $s_i^2$ è¡¨ç¤ºç¬¬$i$ç»„æ•°æ®çš„æ ·æœ¬æ–¹å·®ã€‚

æ³¨æ„åˆ°ï¼Œå¦‚æœæ ·æœ¬æ–¹å·®$s_i^2$å·®å¼‚å¤§ï¼Œé‚£ä¹ˆ$q$éå¸¸å¤§ï¼›å¦‚æœæ ·æœ¬æ–¹å·®$s_i^2$å®Œå…¨ä¸€è‡´ï¼Œé‚£ä¹ˆ$q$ä¸ºé›¶ã€‚åŸºäºè¿™ä¸ªæƒ³æ³•ï¼Œå¦‚æœæ£€éªŒç»Ÿè®¡é‡$\chi_0^2$è¿‡å¤§ï¼Œé‚£ä¹ˆæˆ‘ä»¬å°±æ‹’ç»åŸå‡è®¾ã€‚å…·ä½“æ¥è¯´ï¼Œæˆ‘ä»¬æ‹’ç»åŸå‡è®¾å½“ä¸”ä»…å½“
$$
\chi_0^2 > \chi_{1-\alpha,a-1}^2
$$
å…¶ä¸­$\chi_{1-\alpha,a-1}^2$æ˜¯è‡ªç”±åº¦ä¸º$a-1$çš„å¡æ–¹åˆ†å¸ƒçš„$1-\alpha$åˆ†ä½æ•°ã€‚

ç„¶è€Œï¼ŒBartlettæ£€éªŒå­˜åœ¨ä¸€ä¸ª**ç¼ºé™·**ï¼šBartlettæ£€éªŒå¯¹æ­£æ€åˆ†å¸ƒå‡å®šéå¸¸æ•æ„Ÿã€‚å¦‚æœæ•°æ®èƒŒç¦»æ­£æ€æ€§å‡å®šï¼Œå»ºè®®ä¸è¦ä½¿ç”¨Bartlettæ£€éªŒã€‚

````python
# Bartlett's Test
Bart_stat, Bart_pVal = stats.bartlett(Group3_data[0], Group3_data[1], Group3_data[2])
print("Bartlett's test statistic is", round(Bart_stat,4))
print("The p value is", round(Bart_pVal,4))
````

y_{ij}^{\ast} = |y_{ij} - \tilde{y}_{i}|, i = 1,2,\cdots,a; j = 1,2,\cdots,n_i.
$$
å…¶ä¸­ï¼Œ$\tilde{y}_{i}$è¡¨ç¤ºç¬¬$i$ç»„æ•°æ®çš„ä¸­ä½æ•°ã€‚åŸºäº$\{y_{ij}^{\ast}\}$æ¥æ„å»º$F$ç»Ÿè®¡é‡ã€‚è¿™ä¸ªç»Ÿè®¡é‡å¯ä»¥ç”¨æ¥è¯„ä»·æ¯ç»„æ•°æ®çš„ç»å¯¹åå·®çš„å‡å€¼æ˜¯å¦ç›¸ç­‰ï¼Œä»è€Œå¯ä»¥åˆ¤æ–­æ¯ç»„æ•°æ®çš„æ³¢åŠ¨æ˜¯å¦ä¸€è‡´ã€‚

##### Shapiro-Wilk æ£€éªŒ

å‡è®¾æ•°æ®$\{x_1,x_2,\cdots,x_n\}$æ¥è‡ªæ­£æ€æ€»ä½“$N(\mu,\sigma^2)$ã€‚ä»¤$x_{(1)}\leq x_{(2)}\leq \cdots \leq x_{(n)}$ä¸ºå…¶æ¬¡åºç»Ÿè®¡é‡ã€‚åœ¨Shapiro-Wilk æ£€éªŒä¸­ï¼Œæ£€éªŒç»Ÿè®¡é‡ä¸º
$$
W = \frac{\left(\sum_{i=1}^n (a_{i}-\bar{a}) (x_{i}-\bar{x})\right)^2}{\sum_{i=1}^n (a_i-\bar{a})^2 \sum_{i=1}^n (x_{(i)} - \bar{x})^2}
$$
å…¶ä¸­$a_1,a_2,\cdots,a_n$åœ¨ç»™å®šæ ·æœ¬é‡$n$æ—¶æ˜¯ä¸€äº›ç‰¹å®šçš„å€¼ã€‚è¿™é‡Œå¯ä»¥è¿‘ä¼¼çš„çœ‹ä½œâ€œç†è®ºåˆ†ä½æ•°â€ã€‚äºæ˜¯ï¼Œæ£€éªŒç»Ÿè®¡é‡$W$å¯ä»¥è®¤ä¸ºæ˜¯ç†è®ºåˆ†ä½æ•°å’Œæ ·æœ¬åˆ†ä½æ•°çš„ç›¸å…³ç³»æ•°çš„å¹³æ–¹ã€‚å¦‚æœå®ƒä»¬çº¦æ¥è¿‘äºä¸€æ¡ç›´çº¿ï¼Œé‚£ä¹ˆæˆ‘ä»¬å°±è®¤ä¸ºæ•°æ®æ˜¯æœä»æ­£æ€åˆ†å¸ƒçš„ã€‚äºæ˜¯ï¼Œæˆ‘ä»¬æ‰€æ„é€ çš„æ‹’ç»åŸŸæ˜¯$\{W\leq W_{\alpha}\}$ã€‚åœ¨å®é™…æ•°æ®åˆ†æä¸­ï¼Œæˆ‘ä»¬å¯ä»¥ç”¨pythonä»£ç æ¥å®ç°ã€‚

````python
SW_stat,SW_pVal = stats.shapiro(Res1)
print("Shapiro-Wilk test statistic is", round(SW_stat,4))
print("The p value is", round(SW_pVal,4))
````

## Task 5: è¯•ç”¨Tukeyæ–¹æ³•æ¯”è¾ƒ 3 ä¸ªé¥®é£Ÿæ–¹å¼æ°´å¹³ä¸‹çš„å‡é‡ç¨‹åº¦ä¸¤ä¸¤æ˜¯å¦å­˜åœ¨å·®å¼‚ã€‚
æ¥ä¸‹æ¥ï¼Œæˆ‘ä»¬éœ€è¦åˆ¤æ–­å“ªå‡ å¯¹æ˜¯å­˜åœ¨å·®å¼‚çš„ï¼Œå¹¶ä¸”éœ€è¦é‡åŒ–è¿™äº›å·®å¼‚æ˜¯å¦æ˜¯æ˜¾è‘—çš„ï¼Ÿè¿™é‡Œæˆ‘ä»¬é‡‡ç”¨çš„æ˜¯ç†è®ºè¯¾ç¨‹é‡Œä»‹ç»è¿‡çš„æ–¹æ³•â€”â€”Tukeyã€‚åœ¨Tukeyæ–¹æ³•ä¸­ï¼Œéœ€è¦è®¡ç®—æ¯ç»„æ ·æœ¬å‡å€¼çš„å·®ï¼Œå¹¶å°†å…¶ä¸ä¸´ç•Œå€¼
$$
c = q_{1-\alpha}(a,df) \cdot \hat{\sigma}/\sqrt{m}.
$$
è¿›è¡Œæ¯”è¾ƒã€‚è¿™é‡Œæ‰€è®¡ç®—çš„åˆ†ä½æ•°ï¼Œå®é™…ä¸Šæ˜¯$t$åŒ–æå·®ç»Ÿè®¡é‡çš„åˆ†ä½æ•°ï¼Œå…¶ä¸­$t$åŒ–æå·®ç»Ÿè®¡é‡å®šä¹‰ä¸º
$$
q(a,df) = \max_i \frac{\bar{y}_{i\cdot} - \mu}{\hat{\sigma}/\sqrt{m}}  - \min_i \frac{\bar{y}_{i\cdot} - \mu}{\hat{\sigma}/\sqrt{m}}.
$$
è¿™é‡Œ$a$è¡¨ç¤ºå› å­æ°´å¹³æ•°ï¼Œ$m$è¡¨ç¤ºä¸€ç»„å†…çš„æ•°æ®ä¸ªæ•°ï¼Œ$\hat{\sigma}$æ˜¯æ ‡å‡†å·®çš„ä¼°è®¡ï¼Œå³$\hat{\sigma}=\sqrt{\hat{\sigma}^2}$ã€‚

å€¼å¾—æ³¨æ„çš„æ˜¯ï¼Œè¿™é‡Œ$q_{1-\alpha}(a,df)$è¡¨ç¤º$t$åŒ–æå·®ç»Ÿè®¡é‡çš„åˆ†ä½æ•°ï¼Œé€šå¸¸éœ€è¦æŸ¥è¡¨è€Œå¾—ã€‚![Quantiles_T2.jpeg](Figures/Quantiles_T2.jpeg)

- å¦‚æœä¸¤ç»„æ ·æœ¬å‡å€¼çš„ç»å¯¹å·®å°äºç­‰äºä¸´ç•Œå€¼ï¼Œé‚£ä¹ˆè®¤ä¸ºè¿™ä¸¤ç»„æ ·æœ¬çš„å‡å€¼æ˜¯ç›¸åŒçš„ã€‚
